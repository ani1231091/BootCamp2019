```python
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
titanic=data ("Titanic")
```

```python
trees = data('trees')
sleepstudy = data('sleepstudy')
road = data('road')
birthwt = data('birthwt')
lung = data('lung')
```

```python
trees.plot(kind='scatter',x='Girth', y='Volume', title='Volume/Girth on Black Cherry Trees Felled in 1976')
plt.xlabel('Tree Diameter (inches)')
plt.ylabel('Tree Volume (cubic ft.)')
#print(r"Source: Ryan, T. A., Joiner, B. L. and Ryan, B. F. (1976) _The Minitab Student Handbook_. Duxbury Press.")


trees.plot(kind='scatter',x='Height', y='Volume', title='Volume/Height on Black Cherry Trees Felled in 1976')
plt.xlabel('Tree Height (ft.)')
plt.ylabel('Tree Volume (cubic ft.)')
print(r"Source: Ryan, T. A., Joiner, B. L. and Ryan, B. F. (1976) _The Minitab Student Handbook_. Duxbury Press.")

print('\nTrees Dataset: \nSeems like diameter is a better predictor of volume than Height')
```

```python
sleepstudy.plot(kind='Hexbin',gridsize=20,
                x='Subject', y='Reaction')
plt.xlabel('Subject identifier')
plt.ylabel('Reaction time (ms)')
plt.title('Reaction time by person')
plt.tight_layout()

sleepstudy.plot(kind='Hexbin',gridsize=20,
                x='Days', y='Reaction')
plt.xlabel('Days sleep deprived (days)')
plt.ylabel('Reaction time (ms)')
plt.title('Reaction time by day sleep deprived')
plt.tight_layout()

sleepstudy.plot(kind='box', y='Reaction', vert=False)
plt.title('Reaction times')
plt.xlabel('Reaction time (ms)')
```

```python
road.plot(kind='scatter', x='drivers', y='deaths')
plt.title('Car deaths compared to number of drivers')
plt.xlabel('Number of drives (10,000s)')
plt.ylabel('Number of deaths')

road.plot(kind='box', y='deaths', vert=False)
plt.xlabel('Death in numbers')
plt.title('Distribution of Deaths among states')
```

```python
islow = birthwt['low'] == 1
lowBabies = birthwt[islow]

lowBabies[['age']].plot(kind='box', vert=False)
plt.title('Distribution of ages of mothers for low-weight babies')
plt.xlabel('Age of mother')

lowBabies[['smoke']].hist('smoke')
plt.title('Mother smoking indicator for low-weight babies')
plt.xlabel('Does mother smoke, 1=yes, 0=no')

plt.show()
print('Source: Hosmer, D.W. and Lemeshow, S. (1989) _Applied Logistic Regression._ New York:Wiley')
```
